Importing dependcies

```python
import pandas as pd
import json
import os
```

***Agrreagated transactions***

```python
path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

aggr={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              aggr['Transaction_type'].append(Name)
              aggr['Transaction_count'].append(count)
              aggr['Transaction_amount'].append(amount)
              aggr['State'].append(i)
              aggr['Year'].append(j)
              aggr['Quater'].append(int(k.strip('.json')))

agr_trans_df = pd.DataFrame(aggr)
agr_trans_df


```

```python
agr_trans_df.to_csv('agr_transaction_data.csv')
```

***Agrregated user***

```python
path="/content/pulse/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path)


aggr1={'State':[], 'Year':[],'Quater':[],'brand':[],'count':[], 'percentage':[]}
Agg_state_list
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            
            if D["data"]["usersByDevice"] is None:
                    continue
            for z in D["data"]["usersByDevice"]:
              #print(z)
              brand = z['brand']
              count = z['count']
              percentage = z['percentage']
            
              
              aggr1['brand'].append(brand)
              aggr1['count'].append(count)
              aggr1['percentage'].append(percentage)
              aggr1['State'].append(i)
              aggr1['Year'].append(j)
              aggr1['Quater'].append(int(k.strip('.json')))
              
    
agr_user_df = pd.DataFrame(aggr1)             
agr_user_df  
```

```python
agr_user_df.to_csv('agr_user_data.csv')
```

***Map Transaction***

```python
path="/content/pulse/data/map/transaction/hover/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

total = {'State':[],'Quater':[], 'district':[],'count':[] ,'amount':[],'Year':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']:
              
              for district_data in D['data']["hoverDataList"]:
                  district_name = district_data["name"]
                  district_amount = district_data["metric"][0]["amount"]
                  district_count = district_data["metric"][0]["count"]
                  
                  total['district'].append(district_name)
                  total['count'].append(district_amount)
                  total['amount'].append(district_count)
                  total['State'].append(i)
                  total['Year'].append(j)
                  total['Quater'].append(int(k.strip('.json')))

map_trans_df = pd.DataFrame(total)
map_trans_df

              
```

```python
map_trans_df.to_csv('map_transaction_data.csv')
```

***Map User***

```python
path="/content/pulse/data/map/user/hover/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

map_user = {'State':[],"District": [], "Registered Users": [],'Year':[],'Quater':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D:
              hoverData = D["data"]["hoverData"]

              for district, data in hoverData.items():
                  map_user['District'].append(district)
                  map_user['Registered Users'].append(data["registeredUsers"])
                  map_user['State'].append(i)
                  map_user['Year'].append(j)
                  map_user['Quater'].append(int(k.strip('.json')))
      
map_user_df = pd.DataFrame(map_user)
map_user_df 
             
```

```python
map_user_df.to_csv('map_user_data.csv')
```

***Top Transaction***

```python
path="/content/pulse/data/top/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list


top = {'State':[], 'Year':[],'Quater':[],'Districts':[], 'count':[], 'amount':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['districts']:
              #d_data = D['data']['districts']
              dis_Name=z['entityName']
              count=z['metric']['count']
              amount=z['metric']['amount']
              top['Districts'].append(dis_Name)
              top['count'].append(count)
              top['amount'].append(amount)
              top['State'].append(i)
              top['Year'].append(j)
              top['Quater'].append(int(k.strip('.json')))             
top_trans_df = pd.DataFrame(top)
top_trans_df                  
```

```python
top_trans_df.to_csv('top_transaction_data.csv')
```
